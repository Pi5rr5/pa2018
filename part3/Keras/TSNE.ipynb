```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
data = np.genfromtxt('inputs_keras.csv', delimiter=",", skip_header=1, dtype=float)
np.random.shuffle(data)
inputs = data[10:, 0:20].astype('float32')
outputs = data[10:, 20:21].astype(int)
```

```python
X_scale = StandardScaler().fit_transform(inputs)
```

```python
tsne = TSNE(n_components=2, init='pca', random_state=0)
X_t = tsne.fit_transform(X_scale)
```

```python
plt.figure()
plt.scatter(X_t[np.where(outputs == 0), 0],
                   X_t[np.where(outputs == 0), 1],
                   marker='x', color='g',
                   linewidth='1', alpha=0.8, label='Male')
plt.scatter(X_t[np.where(outputs == 1), 0],
                   X_t[np.where(outputs == 1), 1],
                   marker='v', color='r',
                   linewidth='1', alpha=0.8, label='Female')
```
